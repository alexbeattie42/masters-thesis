```python
import dill as pickle
import banpei
import matplotlib.pyplot as plt
import pandas as pd
```

## Dataset
Run the preprocessing notebook to generate the dataset

```python
keys = ["f_c","P","I_phaseA","I_phaseB","I_phaseC","V_phaseA","V_phaseB","V_phaseC","fault"]
with open('../#datasets/Weak_Grid-7_5_mH.pkl', 'rb') as file:
    df= pickle.load(file)
df = df[:180000].copy()
df[keys].plot(subplots=True, figsize=(16,15));
```

```python
df_small = df[40000:60000].copy()
# Converting the index as date
df_small.index = pd.to_datetime(df_small.index)
model   = banpei.SST(w=50)
model2 = banpei.Hotelling()
series = df_small["P"]
# print(series)
anom = model.detect(series)
anom2 = model2.detect(series,0.5)

plt.plot(anom)
plt.figure()
plt.plot(anom2)


```
